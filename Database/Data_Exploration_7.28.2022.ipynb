```python
import pandas as pd
import numpy as np
from datetime import datetime

pd.options.display.max_rows = 150
pd.options.display.max_columns = 150
```

```python
# Import Data from Downloaded .csv file
# Create a DataFrame for the US-Accident data.
file_to_load = "../US_Accidents_Dec21_updated.csv"
accidents = pd.read_csv(file_to_load)
accidents.head()
```

```python
# Get Row Count for the accidents dataframe
row_count = len(accidents)
row_count
```

```python
#Add CLEAN DATA header
```

```python
# Redefine columns in dataframe
accidents_updated = accidents[['Severity', 'Start_Time', 'State', 'Temperature(F)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)', 
    'Weather_Condition', 'Sunrise_Sunset']].copy()
accidents_updated.head()
```

```python
# Remove NaN values
accidents_updated = accidents_updated.dropna(axis=0, how="any")
accidents_updated.head()
```

```python
# Get row count of accidents_updated dadtaframe
row_count = len(accidents_updated)
row_count
```

```python
# Add 'year' index column from 'Start_Time' column
accidents_updated['year'] = pd.DatetimeIndex(accidents_updated['Start_Time']).year
accidents_updated.head()
```

```python
# Add 'date' index column from 'Start_Time' column
accidents_updated['date'] = pd.DatetimeIndex(accidents_updated['Start_Time']).date
accidents_updated.head()
```

```python
# Review Dtypes for "date" column
accidents_updated.dtypes
```

```python
# Change 'date' column to intergar
accidents_updated['date'] = pd.to_datetime(accidents_updated['date'])
accidents_updated.dtypes
```

```python
# Drop 'State_Time' column
accidents_updated = accidents_updated.drop('Start_Time', axis='columns')
accidents_updated.head()
```

```python
# Rename the columns for ease of use in SQL
accidents_updated.rename(columns = {'Severity':'severity', 
                                    'State':'state_code', 
                                    'Temperature(F)':'temperature_f', 
                                    'Visibility(mi)': 'visibility_mi',
                                    'Wind_Speed(mph)': 'wind_speed_mph',
                                    'Precipitation(in)': 'precipitation_in',
                                    'Weather_Condition': 'weather_condition',
                                    'Sunrise_Sunset': 'sunrise_sunset'}, inplace = True)
accidents_updated.head()
```
